```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_LondonHousePrices = "https://data.london.gov.uk/download/uk-house-price-index/70ac0766-8902-4eb5-aab5-01951aaed773/UK%20House%20price%20index.xls"
properties = pd.read_excel(url_LondonHousePrices, sheet_name='Average price', index_col= None)
```

```python
#Transpose table
properties_T = properties.transpose()
```

```python
properties_T.head(10)
```

```python
#reset index
properties_T = properties_T.reset_index()
properties_T.head()
```

```python
#change column names to first row and drop first row
properties_T.columns = properties_T.iloc[0]
properties_T = properties_T.drop(0)
```

```python
properties_T.head()
```

```python
#rename 1st 2 columns
properties_T = properties_T.rename(columns = {'Unnamed: 0':'London_Borough', pd.NaT: 'ID'})
properties_T.head()
```

```python
#melt table to reduce number of columns
properties_TM = pd.melt(properties_T, id_vars= ['London_Borough', 'ID'])
properties_TM.head()
```

```python
#rename columns
properties_TM = properties_TM.rename(columns = {0: 'Month', 'value': 'Average_price'})
properties_TM.head(10)
```

```python
#change average price to float type
properties_TM['Average_price']=pd.to_numeric(properties_TM['Average_price'])
properties_TM.dtypes
```

```python
properties_TM.count()
```

```python
properties_TM['London_Borough'].unique()
```

```python
properties_TM[properties_TM['London_Borough'] == 'Unnamed: 34'].head()
```

```python
properties_TM[properties_TM['London_Borough'] == 'Unnamed: 37'].head()
```

```python
#remove rows with NaN in ID
values_only = properties_TM[properties_TM['ID'].notna()]
values_only.head(50)
```

```python
values_only.count()
```

```python
clean_prop = properties_TM.dropna()
clean_prop.count()
```

```python
clean_prop['London_Borough'].unique()
```

```python
no_boroughs = ['Inner London', 'Outer London', 
               'NORTH EAST', 'NORTH WEST', 'YORKS & THE HUMBER', 
               'EAST MIDLANDS', 'WEST MIDLANDS',
              'EAST OF ENGLAND', 'LONDON', 'SOUTH EAST', 
              'SOUTH WEST', 'England']
```

```python
#remove values in London_Borough that are not boroughs of London
clean_prop2 = clean_prop[~clean_prop["London_Borough"].isin(no_boroughs)]
clean_prop2['London_Borough'].unique()
```

```python
#clean dataframe name, change to df for simplicity
df = clean_prop2
df.dtypes
```

```python
df.head()
```

```python
#visualization
#restricted to Camden
camden = df[df['London_Borough']=='Camden']
```

```python
grph_camden = camden.plot(kind="line", x='Month', y='Average_price')
grph_camden.set_ylabel("Average Price")
grph_camden.set_xlabel("Month")
```

```python
#use lamda() function to extract year from ['Month'] to reduce data points
df['Year'] = df['Month'].apply(lambda t: t.year)
```

```python
df.tail()
```

```python
df.head()
```

```python
#group by Borough and Year using mean value of average price for the variables
dfg=df.groupby(['London_Borough','Year']).mean()
```

```python
dfg.sample(10)
```

```python
#reset the index 
dfg=dfg.reset_index()
dfg.head()
```

```python
dfg.tail()
```

```python
# 3. Modeling
#function to compare 1998 and 2018 prices for each borough
def price_comparison(d):
    year1998 = float(d['Average_price'][d['Year']==1998])
    year2018 = float(d['Average_price'][d['Year']==2018])
    ratio = [((year2018-year1998)/year1998)*100]
    return ratio
```

```python
price_comparison(dfg[dfg['London_Borough']=='Camden'])
```

```python
#iterate though dfg for each borough to get comparison of 1998/2018 for each borough
final = {}
for b in dfg['London_Borough'].unique():
    borough = dfg[dfg["London_Borough"] == b]
    final[b] = price_comparison(borough)
print(final)
```

```python
df_ratios = pd.DataFrame(final)
```

```python
df_ratios.head()
```

```python
df_ratios_T = df_ratios.transpose()
```

```python
df_ratios_T=df_ratios_T.reset_index()
df_ratios_T.head()
```

```python
df_ratios_T.rename(columns={'index':'Borough', 0 : 'Pct Change'}, inplace=True)
df_ratios_T.head()
```

```python
top10 = df_ratios_T.sort_values(by = 'Pct Change', ascending = False).head(10)
print(top10)
```

```python
#plot top10 
grph_top10=top10[['Borough','Pct Change']].plot(kind='bar')
grph_top10.set_xticklabels(top10.Borough)
```
